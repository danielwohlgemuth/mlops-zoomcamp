**Q6. Evaluating the model**

```python
def get_dataframe(filename):
    df = pd.read_parquet(filename)

    df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
    df.duration = df.duration.apply(lambda td: td.total_seconds()/60)
    duration_1_to_60_minutes = (df.duration >= 1) & (df.duration <= 60)
    df = df[duration_1_to_60_minutes]

    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)

    return df
```

```python
df_val = get_dataframe('yellow_tripdata_2023-02.parquet')
```

```python
def predict(df, dv, lr):
    categorical = ['PULocationID', 'DOLocationID']
    numerical = ['duration']
    train_dicts = df[categorical + numerical].to_dict(orient='records')
    X_train = dv.transform(train_dicts)
    y_pred = lr.predict(X_train)
    return y_pred
```

```python
y_pred_val = predict(df_val, dv, lr)
```

```python
y_train_val = df_val[target].values
```

```python
root_mean_squared_error(y_train_val, y_pred_val)
```

```python
format(root_mean_squared_error(y_train_val, y_pred_val), 'f')
```

What's the RMSE on validation?<br>
0.000020 (differs from the expected values of 3.81, 7.81, 11.81, an 16.81)
